```python
# Instalación de paquetes necesarios (ejecutar primero)
!pip install pandas matplotlib seaborn plotly

# Importar librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from datetime import datetime

# Configuración de visualización
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12, 6)
sns.set_palette("husl")

# 1. Cargar datos (reemplaza con tu archivo)
from google.colab import files
uploaded = files.upload()

# Cargar el dataset
import io
df = pd.read_csv(io.BytesIO(uploaded[next(iter(uploaded))]))

# 2. Procesamiento de datos para sectores
sectores = {
    'cf_comercial_publico': 'CF Comercial & Publico',
    'cf_construcciones': 'CF Construcciones',
    'cf_industrial': 'CF Industrial'
}

consumo_sectores = pd.DataFrame()
for codigo, nombre in sectores.items():
    temp_df = df[df['tipo_energetico'].str.contains(nombre, na=False)]
    if not temp_df.empty:
        consumo_sectores[codigo] = temp_df['con_final']

# 3. Distribución mensual
consumo_mensual = consumo_sectores.copy()
for col in consumo_mensual.columns:
    consumo_mensual[col] = consumo_mensual[col] / 12  # Distribución mensual

meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun',
         'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
consumo_mensual['Mes'] = meses
consumo_mensual.set_index('Mes', inplace=True)

# 4. Visualización con Plotly (mejor para Colab)
fig = px.bar(consumo_mensual,
             title='Consumo Energético Mensual por Sector - Colombia 2015',
             labels={'value': 'Consumo', 'variable': 'Sector'},
             barmode='stack')

fig.update_layout(
    xaxis_title='Mes',
    yaxis_title='Consumo (Unidades Originales)',
    legend_title='Sector'
)

fig.show()

# 5. Análisis de participación
total_anual = consumo_sectores.sum()
participacion = (total_anual / total_anual.sum()) * 100

fig_pie = px.pie(values=participacion,
                 names=participacion.index,
                 title='Participación por Sector en el Consumo Energético 2015')
fig_pie.show()

# 6. Exportar resultados
consumo_mensual.to_csv('consumo_mensual_sectores_2015.csv')
files.download('consumo_mensual_sectores_2015.csv')

# Informe resumido
print("\n🔍 RESUMEN ANUAL:")
print(total_anual.to_markdown())
print("\n📊 PARTICIPACIÓN POR SECTOR:")
print(participacion.to_markdown())
```

```python
import pandas as pd
from google.colab import files

# 1. Subir el archivo
uploaded = files.upload()

# 2. Detectar codificación automáticamente
def detectar_encoding(archivo):
    with open(archivo, 'rb') as f:
        rawdata = f.read(10000)  # Lee solo los primeros 10kb para análisis
    encodings = ['utf-8', 'latin1', 'ISO-8859-1', 'cp1252', 'utf-16']
    for enc in encodings:
        try:
            rawdata.decode(enc)
            return enc
        except UnicodeDecodeError:
            continue
    return 'latin1'  # Valor por defecto si no se detecta

# 3. Leer el archivo con la codificación correcta
archivo = list(uploaded.keys())[0]
encoding = detectar_encoding(archivo)

try:
    df = pd.read_csv(archivo, encoding=encoding)
    print(f"✅ Archivo cargado con codificación: {encoding}")
    print(df.head())
except Exception as e:
    print(f"❌ Error: {e}")
```

```python
import pandas as pd
import numpy as np

# 1. Eliminar columnas totalmente vacías o redundantes
df_clean = df.dropna(axis=1, how='all')

# 2. Conversión de notación científica a decimales
cols_numericas = df_clean.select_dtypes(include=[np.number]).columns
df_clean[cols_numericas] = df_clean[cols_numericas].apply(lambda x: round(x, 2))

# 3. Normalización de unidades (ejemplo para TJ)
if 'Unidades Originales' in df_clean.columns:
    df_clean['Consumo_TJ'] = df_clean.apply(
        lambda row: (
            row['Total Carretero'] if row['Unidades Originales'] == 'TJ' else
            row['Total Carretero'] * 0.000277778 if row['Unidades Originales'] == 'TCal' else
            row['Total Carretero'] * 0.0036  # Conversión Mpc → TJ
        ), axis=1
    )

# 4. Eliminar duplicados temporales
df_clean = df_clean.drop_duplicates(subset=['Anio', 'Tipo Energetico'], keep='last')

# 5. Tratamiento de valores nulos/ceros
cols_ceros = ['Aereo', 'Bunker', 'Perdidas']
df_clean[cols_ceros] = df_clean[cols_ceros].replace(0, np.nan)

# 6. Verificación final
print("\n🔍 Resumen post-limpieza:")
print(f"- Columnas: {len(df_clean.columns)} (Original: {len(df.columns)})")
print(f"- Filas: {len(df_clean)}")
print(f"- Valores nulos: {df_clean.isnull().sum().sum()}")
print("\nMuestra de datos limpios:")
display(df_clean.head(3))
```

```python
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import plotly.express as px

# Configuración de la página
st.set_page_config(page_title="Análisis Energético Colombia", page_icon="⚡", layout="wide")

# Título principal
st.title("📊 Análisis del Balance Energético Colombiano 2015")

# Carga de datos
@st.cache_data
def load_data():
    # Reemplaza con la ruta de tu archivo limpio
    df = pd.read_csv("balance_energetico_colombia_2015_clean.csv")
    return df

df = load_data()

# Sidebar con controles
st.sidebar.header("Opciones de Análisis")
analysis_type = st.sidebar.selectbox(
    "Seleccione el tipo de análisis",
    ["Exploración de Datos", "Correlaciones", "Modelos Predictivos", "Clustering"]
)

# Análisis Exploratorio
if analysis_type == "Exploración de Datos":
    st.header("Exploración Inicial de Datos")

    # Mostrar dataframe
    if st.checkbox("Mostrar datos crudos"):
        st.dataframe(df)

    # Estadísticas descriptivas
    st.subheader("Estadísticas Descriptivas")
    st.dataframe(df.describe())

    # Selección de variable para histograma
    var_hist = st.selectbox("Seleccione variable para histograma", df.columns)
    fig, ax = plt.subplots()
    sns.histplot(df[var_hist], kde=True, ax=ax)
    st.pyplot(fig)

# Análisis de Correlación
elif analysis_type == "Correlaciones":
    st.header("Análisis de Correlaciones")

    # Matriz de correlación
    corr_matrix = df.corr()
    fig, ax = plt.subplots(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, ax=ax)
    st.pyplot(fig)

    # Gráfico de dispersión interactivo
    st.subheader("Gráfico de Dispersión Interactivo")
    col1, col2 = st.columns(2)
    x_axis = col1.selectbox("Eje X", df.columns)
    y_axis = col2.selectbox("Eje Y", df.columns)
    fig = px.scatter(df, x=x_axis, y=y_axis, hover_data=[df.index])
    st.plotly_chart(fig, use_container_width=True)

# Modelos Predictivos
elif analysis_type == "Modelos Predictivos":
    st.header("Modelos de Machine Learning")

    # Selección de características
    st.subheader("Configuración del Modelo")
    features = st.multiselect(
        "Seleccione variables predictoras",
        df.columns,
        default=['oferta_interna_bruta', 'importacion', 'exportaciones']
    )
    target = st.selectbox("Variable objetivo", df.columns)

    if st.button("Entrenar Modelos"):
        # Preparación de datos
        X = df[features]
        y = df[target]

        # Escalado y división
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)
        X_train, X_test, y_train, y_test = train_test_split(
            X_scaled, y, test_size=0.2, random_state=42
        )

        # Entrenamiento de modelos
        col1, col2 = st.columns(2)

        with col1:
            st.subheader("Regresión Lineal")
            lr = LinearRegression()
            lr.fit(X_train, y_train)
            y_pred = lr.predict(X_test)

            st.metric("R²", f"{r2_score(y_test, y_pred):.3f}")
            st.metric("MSE", f"{mean_squared_error(y_test, y_pred):.2e}")

            # Coeficientes
            coeff_df = pd.DataFrame({
                'Variable': features,
                'Coeficiente': lr.coef_
            }).sort_values('Coeficiente', ascending=False)
            st.dataframe(coeff_df)

        with col2:
            st.subheader("Random Forest")
            rf = RandomForestRegressor(n_estimators=100, random_state=42)
            rf.fit(X_train, y_train)
            y_pred_rf = rf.predict(X_test)

            st.metric("R²", f"{r2_score(y_test, y_pred_rf):.3f}")
            st.metric("MSE", f"{mean_squared_error(y_test, y_pred_rf):.2e}")

            # Importancia de características
            importances = pd.DataFrame({
                'Variable': features,
                'Importancia': rf.feature_importances_
            }).sort_values('Importancia', ascending=False)

            fig, ax = plt.subplots()
            sns.barplot(x='Importancia', y='Variable', data=importances, ax=ax)
            st.pyplot(fig)

# Análisis de Clustering
else:
    st.header("Análisis de Clustering")

    # Selección de características para clustering
    cluster_features = st.multiselect(
        "Variables para clustering",
        df.columns,
        default=['oferta_interna_bruta', 'importacion']
    )

    if cluster_features and st.button("Ejecutar Clustering"):
        from sklearn.cluster import KMeans
        from sklearn.decomposition import PCA

        # Escalado
        X = df[cluster_features]
        X_scaled = StandardScaler().fit_transform(X)

        # Reducción de dimensionalidad
        pca = PCA(n_components=2)
        principal_components = pca.fit_transform(X_scaled)

        # Clustering
        kmeans = KMeans(n_clusters=3, random_state=42)
        clusters = kmeans.fit_predict(X_scaled)

        # Visualización
        fig, ax = plt.subplots(figsize=(10, 6))
        scatter = ax.scatter(
            principal_components[:, 0],
            principal_components[:, 1],
            c=clusters,
            cmap='viridis',
            alpha=0.6
        )
        plt.colorbar(scatter)
        ax.set_xlabel('Componente Principal 1')
        ax.set_ylabel('Componente Principal 2')
        ax.set_title('Resultados de Clustering (K=3)')
        st.pyplot(fig)

        # Estadísticas por cluster
        df['Cluster'] = clusters
        st.dataframe(df.groupby('Cluster')[cluster_features].mean())

# Notas finales
st.sidebar.markdown("---")
st.sidebar.info(
    "🔍 Datos del Balance Energético Colombiano 2015\n\n"
    "Herramienta desarrollada para análisis de datos energéticos"
)
```

```python
 Paso 1: Instalar dependencias
!pip install streamlit pandas numpy matplotlib seaborn plotly scikit-learn
!npm install localtunnel

# Paso 2: Crear el archivo app.py
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Configuración de la página
st.set_page_config(page_title="Análisis Energético Colombia", page_icon="⚡", layout="wide")

# Título principal
st.title("📊 Análisis del Balance Energético Colombiano")

# Función para cargar datos
@st.cache_data
def load_data(uploaded_file):
    try:
        if uploaded_file.name.endswith('.csv'):
            df = pd.read_csv(uploaded_file)
        elif uploaded_file.name.endswith('.xlsx'):
            df = pd.read_excel(uploaded_file)
        return df
    except Exception as e:
        st.error(f"Error al cargar el archivo: {e}")
        return None

# Interfaz principal
def main():
    # Carga de archivos
    uploaded_file = st.sidebar.file_uploader("Sube tu archivo (CSV o Excel)", type=["csv", "xlsx"])

    if uploaded_file is not None:
        df = load_data(uploaded_file)

        if df is not None:
            st.sidebar.success("✅ Archivo cargado correctamente!")
            analysis_type = st.sidebar.selectbox(
                "Tipo de análisis",
                ["Exploración", "Correlaciones", "Modelos Predictivos", "Clustering"]
            )

            if analysis_type == "Exploración":
                show_exploration(df)
            elif analysis_type == "Correlaciones":
                show_correlations(df)
            elif analysis_type == "Modelos Predictivos":
                show_models(df)
            else:
                show_clustering(df)

def show_exploration(df):
    st.header("Exploración de Datos")

    # Mostrar datos
    if st.checkbox("Mostrar datos completos"):
        st.dataframe(df)

    # Estadísticas
    st.subheader("Estadísticas Descriptivas")
    st.dataframe(df.describe())

    # Histogramas
    st.subheader("Distribución de Variables")
    col = st.selectbox("Selecciona variable", df.columns)
    fig, ax = plt.subplots()
    sns.histplot(df[col], kde=True, ax=ax)
    st.pyplot(fig)

def show_correlations(df):
    st.header("Análisis de Correlaciones")

    # Matriz de correlación
    st.subheader("Matriz de Correlación")
    fig, ax = plt.subplots(figsize=(10, 8))
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm', center=0, ax=ax)
    st.pyplot(fig)

    # Scatter plot interactivo
    st.subheader("Relación entre Variables")
    col1, col2 = st.columns(2)
    x_axis = col1.selectbox("Eje X", df.columns)
    y_axis = col2.selectbox("Eje Y", df.columns)
    fig = px.scatter(df, x=x_axis, y=y_axis, trendline="ols")
    st.plotly_chart(fig, use_container_width=True)

def show_models(df):
    st.header("Modelos Predictivos")

    # Selección de variables
    features = st.multiselect("Variables predictoras", df.columns)
    target = st.selectbox("Variable objetivo", df.columns)

    if st.button("Entrenar Modelos") and features and target:
        X = df[features]
        y = df[target]

        # Preprocesamiento
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)
        X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

        # Modelos
        col1, col2 = st.columns(2)

        with col1:
            st.subheader("Regresión Lineal")
            lr = LinearRegression()
            lr.fit(X_train, y_train)
            y_pred = lr.predict(X_test)

            st.metric("R²", f"{r2_score(y_test, y_pred):.3f}")
            st.metric("MSE", f"{mean_squared_error(y_test, y_pred):.2e}")

            # Coeficientes
            st.write("Coeficientes:")
            coeffs = pd.DataFrame({"Variable": features, "Coef": lr.coef_})
            st.dataframe(coeffs.sort_values("Coef", ascending=False))

        with col2:
            st.subheader("Random Forest")
            rf = RandomForestRegressor(n_estimators=100, random_state=42)
            rf.fit(X_train, y_train)
            y_pred_rf = rf.predict(X_test)

            st.metric("R²", f"{r2_score(y_test, y_pred_rf):.3f}")
            st.metric("MSE", f"{mean_squared_error(y_test, y_pred_rf):.2e}")

            # Importancia de características
            st.write("Importancia de variables:")
            importances = pd.DataFrame({
                "Variable": features,
                "Importancia": rf.feature_importances_
            }).sort_values("Importancia", ascending=False)
            st.dataframe(importances)

def show_clustering(df):
    st.header("Análisis de Clustering")

    # Selección de variables
    features = st.multiselect("Variables para clustering", df.columns)
    n_clusters = st.slider("Número de clusters", 2, 5, 3)

    if st.button("Ejecutar Clustering") and features:
        X = df[features]
        X_scaled = StandardScaler().fit_transform(X)

        # PCA para visualización
        pca = PCA(n_components=2)
        X_pca = pca.fit_transform(X_scaled)

        # Clustering
        kmeans = KMeans(n_clusters=n_clusters, random_state=42)
        clusters = kmeans.fit_predict(X_scaled)

        # Visualización
        fig, ax = plt.subplots(figsize=(10, 6))
        scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap="viridis", alpha=0.6)
        plt.colorbar(scatter)
        ax.set_xlabel("Componente Principal 1")
        ax.set_ylabel("Componente Principal 2")
        st.pyplot(fig)

        # Resultados
        df_cluster = df.copy()
        df_cluster["Cluster"] = clusters
        st.write("Datos con asignación de clusters:")
        st.dataframe(df_cluster)

if __name__ == "__main__":
    main()

# Paso 3: Ejecutar la aplicación
!streamlit run app.py & npx localtunnel --port 8501
```
